### Primero realizo las transformaciones obligatorias que se piden para el MVP

```python
import pandas as pd
import numpy as np
import ast
import calendar
import locale
```

```python
#Ingesta el dataframe 
data= pd.read_csv("movies_dataset.csv")
data.head()
```

```python
#Rellena por el número 0 los valores nulos de los campos revenue y budget 
data['revenue'].fillna(0, inplace=True)
data['budget'].fillna(0, inplace=True)
#verifica cuantos datos nulos hay
data.isnull().sum()
```

```python
# Elimina las columnas 'video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage'
eliminadas = ['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage']
data.drop(columns=eliminadas, inplace=True)
```

```python
#Verifica que el tipo de dato en fecha es datetime sino la convierte con el metodo to_datetime()
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')
data.info()
```

```python
#Elimina todos los nulos de la columna
data.dropna(subset=['release_date'], inplace=True)
```

```python
#Cambie el formato a las fechas en el formato "AAAA-mm-dd" utilizando el método strftime
data['release_date'] = data['release_date'].dt.strftime('%Y-%m-%d')
data.head()
```

```python
#Agrega otra columna que se llama release_year donde se extraera el año de la fecha de estreno que esta en la columna 'release_date'
data['release_date'] = pd.to_datetime(data['release_date'])
data['release_year'] = data['release_date'].dt.year
```

```python
data.head()
```

### Para desanidar las columnas belong to collection y production_companies decidi hacer transformaciones en el tipo de datos y eliminar todo lo que no sea name ya que lo demas no lo iba a utilizar 

```python
#Convierte los valores de la columna 'belongs_to_collection' en diccionarios utilizando ast.literal_eval() solo si el valor es una cadena (str)

data['belongs_to_collection'] = data['belongs_to_collection'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

#Elimina las claves 'poster_path' y 'backdrop_path' del diccionario, dejando solo la clave 'name'.

data['belongs_to_collection'] = data['belongs_to_collection'].apply(lambda x: {'name': x['name']} if isinstance(x, dict) else x)

#Elimina lo que no es name y tambien borra la palabra 

data['belongs_to_collection'] = data['belongs_to_collection'].apply(lambda x: {'name': x['name']} if isinstance(x, dict) else x)

data['belongs_to_collection'] = data['belongs_to_collection'].apply(lambda x: x['name'] if isinstance(x, dict) else x)
```

```python
# Convierte los valores de la columna 'production_companies' en listas de diccionarios utilizando ast.literal_eval() solo si el valor es una cadena (str)
data['production_companies'] = data['production_companies'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Elimina todas las claves excepto 'name' de cada diccionario en la lista
data['production_companies'] = data['production_companies'].apply(lambda x: [{'name': d['name']} for d in x] if isinstance(x, list) else x)

# Extrae el valor asociado con la clave 'name' de cada diccionario en la lista
data['production_companies'] = data['production_companies'].apply(lambda x: [d['name'] for d in x] if isinstance(x, list) else x)
```

```python
data.head()
```

### Para poder crear la columna de retorno de inversion primero se deben convertir los datos a tipo float porque sino no permite las siguientes transformaciones

```python
# Verifica que los datos de revenue y budget son numericos y sino las convierte a float 
data['revenue'] = pd.to_numeric(data['revenue'], errors='coerce')
data['budget'] = pd.to_numeric(data['budget'], errors='coerce')

# Crea la columna de retorno de inversión
data['return'] = np.where(data['budget'] != 0, data['revenue'] / data['budget'], 0)

'''Reemplaza los valores NaN e infinitos por 0, agregue que deje a 0 si hay un valor infinito ya que
   algunos valores aparecian como inf.'''
data['return'] = data['return'].fillna(0)
data['return'] = data['return'].replace(np.inf, 0)


data.head()
```

```python
#guarda las transformaciones que realice en un csv
data.to_csv('movies.csv', index=False)
```

### Por ultimo genero nuevas transformaciones para crear nuevos dataframes para las funciones 

```python
'''Para poder generar las columnas que me den nombre de mes en Español'''
# Crea una función para obtener el nombre del mes en español
def obtener_nombre_mes(mes):
    return calendar.month_name[mes]

# Crea una nueva columna con el nombre del mes en español
data['nombre_mes'] = data['release_date'].dt.month.map(obtener_nombre_mes)

```

```python
data.head()
```

```python
# Configura la localización en castellano
locale.setlocale(locale.LC_TIME, 'es_ES')

# Crear una nueva columna con el nombre del mes en castellano
data['nombre_mes'] = data['release_date'].dt.month.apply(lambda x: calendar.month_name[x])
```

### Trabajo sobre el dataframe de credits para realizar las transformaciones que me permitan desanidar y trabajar con las columnas actors y directors

```python
# Ingesto el dataframe
cred= pd.read_csv("credits.csv")
```

```python
cred.head()
```

```python

# Convierte la columna "cast" a listas de diccionarios
cred['cast'] = cred['cast'].apply(ast.literal_eval)

# Aplica la función lambda para extraer los nombres de los actores
cred['actors'] = cred['cast'].apply(lambda x: [actor['name'] for actor in x])

# Verifica el resultado
print(cred['actors'])
```

```python
#Convierte la columna "crew" en listas de diccionarios utilizando la función ast.literal_eval
import ast
cred['crew'] = cred['crew'].apply(ast.literal_eval)
```

```python
#Utiliza una función lambda junto con una comprensión de lista para filtrar los elementos en los que el valor de la clave "job" sea igual a "Director" y extraer el valor de la clave "name" para esos elementos.
cred['director_names'] = cred['crew'].apply(lambda x: [item['name'] for item in x if item['job'] == 'Director'])
```

```python
#Convierte la lista de nombres de directores en un solo string separado por comas.
cred['director_names'] = cred['director_names'].apply(', '.join)
cred.head()
```

```python
#Crea un nuevo dataframe con los datos de las columnas 
new_cred = cred.loc[:, ['id', 'actors', 'director_names']].copy()
```
